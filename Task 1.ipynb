```python
# filling null values in these columns 
df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Not Available', inplace=True)
df['country'].fillna('Unknown', inplace=True)
df['rating'].fillna('Unrated', inplace=True)
df['duration'].fillna('Unknown', inplace=True)
```

```python
df['date_added']
```

**Standardization:** Stripping extra spaces and capitalizing columns keeping unknowns consistent

```python
df['type'] = df['type'].str.strip().str.title()

df['title'] = df['title'].str.strip()

df['director'] = df['director'].fillna('Unknown').str.strip()

df['cast'] = df['cast'].fillna('Unknown').str.strip()

df['country'] = df['country'].fillna('Unknown').str.strip().str.title()

df['rating'] = df['rating'].fillna('Unknown').str.upper().str.strip()

df['listed_in'] = df['listed_in'].str.strip().str.title()
```

```python
df.head()
```

```python
df.isnull().sum()
```

**Checking data type of each attribute to make necessary changes**

```python
df.dtypes
```

```python
# converting date_added column to datetime dtype

df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['date_added'] = df['date_added'].dt.strftime('%d-%m-%Y')  # format: dd-mm-yyyy
df = df[df['date_added'].notna()] # dropping all 98 invalid entries as they were less compared to the size of the dataset and were not that meaningful.
```

```python
# converting release year column to int

df['release_year'] = df['release_year'].astype('Int64') 
```

```python
#Converting duration column to int

def extract_duration(row):
    if 'min' in row:
        return int(row.split()[0])
    elif 'Season' in row:
        return int(row.split()[0]) 
    return None

df['duration'] = df['duration'].dropna().apply(extract_duration).astype('Int64') 
```

```python
# Insight : Content count by type

sns.countplot(data=df, x='type')
plt.title('Distribution of Movies vs TV Shows')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
# top 5 countries producing Netflix content
top_countries = df['country'].value_counts().head(5)
top_countries.plot(kind='bar', figsize=(10, 5), color='skyblue')
plt.title('Top 5 Content Producing Countries on Netflix')
plt.ylabel('Number of Titles')
plt.xlabel('Country')
plt.xticks(rotation=0)
plt.show()
```
